```python
import os
import numpy as np
import veloxchem as vlx
from MOF_builder.functions.frag_recognizer import process_linker_molecule
from MOF_builder.functions.add_dummy2node import add_dummy_atoms_nodecif
from MOF_builder.gmxmd_prepare.fetchfile import fetch_file,read_mof_top_dict,find_data_folder,copy_file
from MOF_builder.multitopic import multitopic
```

```python
mof_name = 'PCN-222'#'UiO-66'#'PCN-222' #'TEST-MOF'#
#mof_name = 'UiO-66'
metal = 'Zr'
supercell= [1,1,1]
#linker_file='../ndi.xyz'
linker_file = 'CoTCPP.xyz'#'../outndi.xyz'
DUMMY_NODE=False
node_termination = 'methyl' 
sol_list=['tip3p','CO2']
sol_num = [1000,10]
model_name = 'TEST'

#######################################################################
! rm nodes/.DS_Store
! rm nodes/*tri*.cif
! rm edges/.DS_Store
! rm edges/*.cif
supercell = np.asarray(supercell)
mof_top_dict = read_mof_top_dict()
node_connection= mof_top_dict[mof_name]['node_connection']
linker_topic = mof_top_dict[mof_name]['linker_topic']
template = mof_top_dict[mof_name]['topology']+'.cif'
if node_termination == 'methyl' :
    n_term_file = '../data/terminations_database/methyl.pdb'

data_path = find_data_folder()
nodes_database_path = os.path.join(data_path,"nodes_database")
node_cif= fetch_file(nodes_database_path,[str(node_connection)+'c',metal], ['dummy'])[0]

node_cif_database = os.path.join(data_path,'nodes_database/'+node_cif)
target_node_path = os.path.join('nodes',node_cif)
copy_file(node_cif_database,target_node_path)
if DUMMY_NODE:
    add_dummy_atoms_nodecif(target_node_path,metal)
if DUMMY_NODE:
    os.remove(target_node_path)

molecule = vlx.Molecule.read_xyz_file(linker_file)
center_frag_nodes_num,center_Xs,single_frag_nodes_num,frag_Xs= process_linker_molecule(molecule,linker_topic)

templates_dir = '../data/template_database' #default 
nodes_dir = os.path.dirname(target_node_path)#'nodes' #default 
edges_dir = 'edges' #default 

if DUMMY_NODE:
    keywords = [str(node_connection)+'c',metal,'dummy']
    nokeywords = []
else:
    keywords = [str(node_connection)+'c',metal]
    nokeywords = ['dummy']


dummy_node_name = fetch_file(nodes_dir,keywords, nokeywords)[0]
linker_file_ff='/Users/chenxili/GitHub/MOFbuilder/tests/NDI.itp'
#linker_file_ff = ff_gen_xyz(linker_file,charge=-1*linker_topic)
```

```python
new_mof = multitopic(templates_dir,nodes_dir,edges_dir,template,node_connection,linker_topic)
new_mof.load(True)
```

```python
new_mof.basic_supercell(supercell,term_file = n_term_file,boundary_cut_buffer=0.0,edge_center_check_buffer=0.05)
new_mof.write_basic_supercell('31.gro','32.xyz')
```

```python
#cleave all unsaturated linkers
new_mof.defect_missing()
new_mof.term_defective_model(n_term_file,e_termfile = '../data/terminations_database/CCO2.pdb')
new_mof.write_tntemof('30.gro')    
  
'''cleave all unsaturated linkers'''
u_edge_idx=[]
for i in new_mof.unsaturated_main_frag_edges:
    u_edge_idx.append(int(i[0][1:]))

new_mof.defect_missing([],u_edge_idx)
new_mof.term_defective_model(n_term_file,e_termfile = '../data/terminations_database/CCO2.pdb')
new_mof.write_tntemof('33.gro')               
```

```python
'''generate forcefield file or fetch forcefield file from database'''

from MOF_builder.functions.process_gro import split_dummy_node_tntte_arr
from MOF_builder.gmxmd_prepare.forcefield import get_residues_forcefield
arr,node_split_dict = split_dummy_node_tntte_arr(new_mof.tn_te_cc,dummy_node_name,DUMMY_NODE)
new_arr,res_info,restypes=get_residues_forcefield(arr,node_split_dict,DUMMY_NODE,linker_file,linker_file_ff,linker_topic,center_frag_nodes_num,center_Xs,single_frag_nodes_num,frag_Xs)
```

```python
'''copy grofile and corresponding itps to itps subfolder'''

from MOF_builder.gmxmd_prepare.gro_itps import get_gro,get_itps
gro_path=get_gro(model_name,new_arr)
itp_dir=get_itps(data_path,restypes,metal,node_termination,sol_list)
```

```python
'''combine itps for topolgy file'''

from MOF_builder.gmxmd_prepare.top_combine import genrate_top_file
top_path=genrate_top_file(itp_dir,data_path,res_info,model_name)
```

```python
'''copy mdp folders'''

from MOF_builder.gmxmd_prepare.mdps import copy_mdps
mdp_dir=copy_mdps(data_path)
```

```python
'''solvate model and update .top file'''

import subprocess
from MOF_builder.gmxmd_prepare.solvate import solvate_model

init_gro=gro_path
for i in range(len(sol_list)):
    EXECUTE=False
    sv=solvate_model(data_path)
    sv.set_maxsol(int(sol_num[i]))
    sv.set_solvent(sol_list[i])
    sv.set_grofile(init_gro)
    sv.set_top_file(top_path)
    command=sv.solvate()
    if EXECUTE:
        result = subprocess.run(command, shell=True, capture_output=True, text=True)
        # Display the result
        #print("STDOUT:", result.stdout)
        #print("STDERR:", result.stderr)
    init_gro=sv.output_gro
    sv= None
```
